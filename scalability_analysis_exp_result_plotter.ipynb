```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fn = "exp_runners/results/"
methods = ["aupr", "auroc_no_tie", "auroc_with_tie"]
num_repetitions = 5
```

# AUPR

## Delta in AUPR

```python
exp_type = "delta"
tmp_fn_h = fn + methods[0] + "/" + methods[0] + "_" + exp_type + "_helper"
tmp_fn_p1 = fn + methods[0] + "/" + methods[0] + "_" + exp_type + "_p1"
tmp_fn_p2 = fn + methods[0] + "/" + methods[0] + "_" + exp_type + "_p2"
print(tmp_fn_h)
print(tmp_fn_p1)
print(tmp_fn_p2)
deltas = np.array([3, 5, 11, 25, 51, 101])
n_deltas = len(deltas)
exe_time = np.empty(3, n_deltas, num_repetitions, dtype = double)
send_byte = np.empty(3, n_deltas, num_repetitions, dtype = double)
rec_byte = np.empty(3, n_deltas, num_repetitions, dtype = double)

# helper
ind = 0
p_cnt = 0
with open(tmp_fn_h) as f:
    while True:
        line = f.readline()
        if not line:
            break
        words = line.strip().split("\t");
        if words[0] == "Paper":
            send_byte[0][p_cnt][ind] = float(words[1])
            rec_byte[0][p_cnt][ind] = float(words[2])
            exe_time[0][p_cnt][ind] = float(words[3])
            p_cnt += 1
            if p_cnt == n_parties:
                ind += 1
                p_cnt = 0

# p1
ind = 0
p_cnt = 0
with open(tmp_fn_p1) as f:
    while True:
        line = f.readline()
        if not line:
            break
        words = line.strip().split("\t");
        if words[0] == "Paper":
            send_byte[1][p_cnt][ind] = float(words[1])
            rec_byte[1][p_cnt][ind] = float(words[2])
            exe_time[1][p_cnt][ind] = float(words[3])
            p_cnt += 1
            if p_cnt == n_parties:
                ind += 1
                p_cnt = 0

# p2
ind = 0
p_cnt = 0
with open(tmp_fn_p2) as f:
    while True:
        line = f.readline()
        if not line:
            break
        words = line.strip().split("\t");
        if words[0] == "Paper":
            send_byte[2][p_cnt][ind] = float(words[1])
            rec_byte[2][p_cnt][ind] = float(words[2])
            exe_time[2][p_cnt][ind] = float(words[3])
            p_cnt += 1
            if p_cnt == n_parties:
                ind += 1
                p_cnt = 0
            
print("Send Byte", send_byte)
print("\nReceive Byte", rec_byte)
print("\nExecution Time", exe_time)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Delta', fontsize = 16)
ax.set_ylabel('Time (s)', fontsize = 16)

# linestyle = {"linestyle":"--", "linewidth":4, "markeredgewidth":5, "elinewidth":5, "capsize":10}
# ax.errorbar(xaxis, mean1, yerr = se1, color="r", **linestyle)
# ax.errorbar(xaxis, mean2, yerr = se2, color="b", **linestyle)
ax.set_ylim([0, np.max(np.max(exe_time[0])) + 50])
ax.boxplot(exe_time[0].T, positions = deltas)
plt.show()
```
